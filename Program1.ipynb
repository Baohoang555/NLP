```python
def clean_text(text):
    """LÃ m sáº¡ch vÄƒn báº£n"""
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
def process_text(text, dictionary):
    """TÃ¡ch tá»« vÃ  lá»c theo tá»« Ä‘iá»ƒn"""
    clean = clean_text(text)
    
    # TÃ¡ch tá»« vá»›i underthesea
    try:
        underthesea_tokens = underthesea_tokenize(clean)
    except Exception as e:
        print(f"Lá»—i underthesea: {e}")
        underthesea_tokens = clean.split()
    
    # Lá»c tá»« Ä‘iá»ƒn (náº¿u tá»« Ä‘iá»ƒn khÃ´ng trá»‘ng)
    if dictionary:
        underthesea_filtered = [w for w in underthesea_tokens if w in dictionary]
    else:
        underthesea_filtered = underthesea_tokens
    
    # TÃ¡ch tá»« vá»›i pyvn
    try:
        pyvn_tokens = pyvn.tokenize(clean)
    except Exception as e:
        print(f"Lá»—i pyvn: {e}")
        pyvn_tokens = clean.split()
    
    # Lá»c tá»« Ä‘iá»ƒn
    if dictionary:
        pyvn_filtered = [w for w in pyvn_tokens if w in dictionary]
    else:
        pyvn_filtered = pyvn_tokens
    
    return underthesea_tokens, underthesea_filtered, pyvn_tokens, pyvn_filtered
```

```python
def analyze_results(text_underthesea, text_pyvn, 
                   filtered_underthesea, filtered_pyvn):
    """Thá»‘ng kÃª chi tiáº¿t káº¿t quáº£ tÃ¡ch tá»«"""
    
    print("\n" + "=" * 70)
    print("THá»NG KÃŠ Káº¾T QUáº¢ TÃCH Tá»ª")
    print("=" * 70)
    
    # Thá»‘ng kÃª underthesea
    print("\nğŸ“Œ UNDERTHESEA:")
    print(f"  â€¢ Tá»•ng tá»« tÃ¡ch Ä‘Æ°á»£c: {len(text_underthesea)}")
    print(f"  â€¢ Tá»« sau lá»c tá»« Ä‘iá»ƒn: {len(filtered_underthesea)}")
    if len(text_underthesea) > 0:
        ratio = len(filtered_underthesea) / len(text_underthesea) * 100
        print(f"  â€¢ Tá»· lá»‡ tá»« Ä‘Ãºng: {ratio:.2f}%")
    print(f"  â€¢ Máº«u tá»« tÃ¡ch: {' '.join(filtered_underthesea[:10])}...")
    
    # Thá»‘ng kÃª pyvn
    print("\nğŸ“Œ PYVN:")
    print(f"  â€¢ Tá»•ng tá»« tÃ¡ch Ä‘Æ°á»£c: {len(text_pyvn)}")
    print(f"  â€¢ Tá»« sau lá»c tá»« Ä‘iá»ƒn: {len(filtered_pyvn)}")
    if len(text_pyvn) > 0:
        ratio = len(filtered_pyvn) / len(text_pyvn) * 100
        print(f"  â€¢ Tá»· lá»‡ tá»« Ä‘Ãºng: {ratio:.2f}%")
    print(f"  â€¢ Máº«u tá»« tÃ¡ch: {' '.join(filtered_pyvn[:10])}...")
    
    # So sÃ¡nh
    print("\nğŸ“Š SO SÃNH HIá»†U NÄ‚NG:")
    common = set(filtered_underthesea) & set(filtered_pyvn)
    only_underthesea = set(filtered_underthesea) - set(filtered_pyvn)
    only_pyvn = set(filtered_pyvn) - set(filtered_underthesea)
    
    print(f"  â€¢ Tá»« chung giá»¯a 2 thÆ° viá»‡n: {len(common)}")
    print(f"  â€¢ Chá»‰ underthesea: {len(only_underthesea)}")
    print(f"  â€¢ Chá»‰ pyvn: {len(only_pyvn)}")
    
    if only_underthesea:
        print(f"    â†’ VÃ­ dá»¥: {', '.join(list(only_underthesea)[:5])}")
    if only_pyvn:
        print(f"    â†’ VÃ­ dá»¥: {', '.join(list(only_pyvn)[:5])}")
    
    # Táº§n suáº¥t tá»«
    print("\nğŸ”¤ Tá»ª PHá»” BIáº¾N NHáº¤T (Top 10):")
    combined = Counter(filtered_underthesea + filtered_pyvn)
    for i, (word, count) in enumerate(combined.most_common(10), 1):
        print(f"  {i:2d}. {word:15s} : {count:3d} láº§n")
```

```python
def analyze_bigrams(tokens):
    """PhÃ¢n tÃ­ch 2-tá»« (bigrams)"""
    bigrams = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1)]
    bigram_counter = Counter(bigrams)
    return bigram_counter
```
