```python
def clean_text(text):
    """Làm sạch văn bản"""
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
def process_text(text, dictionary):
    """Tách từ và lọc theo từ điển"""
    clean = clean_text(text)
    
    # Tách từ với underthesea
    try:
        underthesea_tokens = underthesea_tokenize(clean)
    except Exception as e:
        print(f"Lỗi underthesea: {e}")
        underthesea_tokens = clean.split()
    
    # Lọc từ điển (nếu từ điển không trống)
    if dictionary:
        underthesea_filtered = [w for w in underthesea_tokens if w in dictionary]
    else:
        underthesea_filtered = underthesea_tokens
    
    # Tách từ với pyvn
    try:
        pyvn_tokens = pyvn.tokenize(clean)
    except Exception as e:
        print(f"Lỗi pyvn: {e}")
        pyvn_tokens = clean.split()
    
    # Lọc từ điển
    if dictionary:
        pyvn_filtered = [w for w in pyvn_tokens if w in dictionary]
    else:
        pyvn_filtered = pyvn_tokens
    
    return underthesea_tokens, underthesea_filtered, pyvn_tokens, pyvn_filtered
```

```python
def analyze_results(text_underthesea, text_pyvn, 
                   filtered_underthesea, filtered_pyvn):
    """Thống kê chi tiết kết quả tách từ"""
    
    print("\n" + "=" * 70)
    print("THỐNG KÊ KẾT QUẢ TÁCH TỪ")
    print("=" * 70)
    
    # Thống kê underthesea
    print("\n📌 UNDERTHESEA:")
    print(f"  • Tổng từ tách được: {len(text_underthesea)}")
    print(f"  • Từ sau lọc từ điển: {len(filtered_underthesea)}")
    if len(text_underthesea) > 0:
        ratio = len(filtered_underthesea) / len(text_underthesea) * 100
        print(f"  • Tỷ lệ từ đúng: {ratio:.2f}%")
    print(f"  • Mẫu từ tách: {' '.join(filtered_underthesea[:10])}...")
    
    # Thống kê pyvn
    print("\n📌 PYVN:")
    print(f"  • Tổng từ tách được: {len(text_pyvn)}")
    print(f"  • Từ sau lọc từ điển: {len(filtered_pyvn)}")
    if len(text_pyvn) > 0:
        ratio = len(filtered_pyvn) / len(text_pyvn) * 100
        print(f"  • Tỷ lệ từ đúng: {ratio:.2f}%")
    print(f"  • Mẫu từ tách: {' '.join(filtered_pyvn[:10])}...")
    
    # So sánh
    print("\n📊 SO SÁNH HIỆU NĂNG:")
    common = set(filtered_underthesea) & set(filtered_pyvn)
    only_underthesea = set(filtered_underthesea) - set(filtered_pyvn)
    only_pyvn = set(filtered_pyvn) - set(filtered_underthesea)
    
    print(f"  • Từ chung giữa 2 thư viện: {len(common)}")
    print(f"  • Chỉ underthesea: {len(only_underthesea)}")
    print(f"  • Chỉ pyvn: {len(only_pyvn)}")
    
    if only_underthesea:
        print(f"    → Ví dụ: {', '.join(list(only_underthesea)[:5])}")
    if only_pyvn:
        print(f"    → Ví dụ: {', '.join(list(only_pyvn)[:5])}")
    
    # Tần suất từ
    print("\n🔤 TỪ PHỔ BIẾN NHẤT (Top 10):")
    combined = Counter(filtered_underthesea + filtered_pyvn)
    for i, (word, count) in enumerate(combined.most_common(10), 1):
        print(f"  {i:2d}. {word:15s} : {count:3d} lần")
```

```python
def analyze_bigrams(tokens):
    """Phân tích 2-từ (bigrams)"""
    bigrams = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1)]
    bigram_counter = Counter(bigrams)
    return bigram_counter
```
